```python
import pdf2image
import pytesseract
from pytesseract import Output, TesseractError
import pandas as pd 
```

```python
pdf_path = "/Users/ismailbhamjee/Documents/hadith nlp/pdf_breakdown/part1.pdf"

images = pdf2image.convert_from_path(pdf_path)
```

```python
df = pd.DataFrame(columns=['text', 'page_number'])
```

```python
text = []
number = []
for x in range(723):
    pil_im = images[x] # assuming that we're interested in the first page only

    ocr_dict = pytesseract.image_to_data(pil_im, lang='eng', output_type=Output.DICT)
    # ocr_dict now holds all the OCR info including text and location on the image

    raw_text = " ".join(ocr_dict['text'])
    print(x)
    
    text.append(raw_text)
    number.append(x)
    #print(x)
    
```

```python
df = pd.DataFrame({'text': text,'page_number': number})
```
