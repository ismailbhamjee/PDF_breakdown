```python
#joins the full chapter into one list
# you can add all the strings to the follwing bookschapters list
book='book_1'
chapter='chapter_1'

df_slice = df2[(df2.book.str.contains(book)) & (df2.chapter.str.contains(chapter))]
x = ' '.join(df_slice['text'].tolist())
books_chapters.append(x)
books_list.append(book)
chapters_list.append(chapter)
len(books_chapters)
```
